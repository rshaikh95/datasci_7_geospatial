```python
import geopandas as  gpd
```

Dataset of interest

Hospitalization Discharge Rates in Lake County, Illinois

https://catalog.data.gov/dataset/public-school-characteristics-current-8d621

```python
df_gpd = gpd.read_file('Hospitalization_Discharge_Rates-shp.zip')
df_gpd
```

```python
df_gpd[['Diabetes', 'ZIP']].sort_values(by=['Diabetes'], ascending=False)
```

Variables chosen:

Diabetes: Diabetes is a chronic disease in which blood sugar (glucose) levels are above normal. This is a rate per 100,000.

Zip

```python

df_gpd['Diabetes'] = (df_gpd['Diabetes'] / 1000)
```

```python
df_gpd.plot("Diabetes", legend=True)
```

Map shows that the rate of diabetes is higher within northern parts of the map. The highest are within the NW portion of map.

Asians in Lake County

```python
df_gpd = gpd.read_file('Birth_Statistics.geojson')
df_gpd
```

Dataset of Interest: Birth Statistics

https://catalog.data.gov/dataset/birth-statistics-a76a6

Births rates across Lake County, Illinois by ZIP Code.

Birth Rate – Birth rate is defined as the number of live births per 1,000 populations.

Zip

```python
df_gpd[['Birth_Rate', 'ZIP']].sort_values(by=['Birth_Rate'], ascending=False)
```

```python
df_gpd['Birth_Rate'] = (df_gpd['Birth_Rate'] / 1000)
```

```python

df_gpd.plot("Birth_Rate", legend=True)
```

Map shows a varied range of birth rate. There are some pockets with high birth rates such as in the right center (42.35, -87.7) and left center (42.35, -88.2). There is a very low birth rate on the south west portion of map.

```python
df_gpd = gpd.read_file('Obesity_Percentages.geojson')
df_gpd
```

Dataset of Interest: Obesity Percentages

https://catalog.data.gov/dataset/obesity-percentages-090d2

Obesity percentages for Lake County, Illinois.
Pct_Obese – The percent of people in the zip code who are considered obese, defined as having a BMI greater than or equal to 30.

```python
df_gpd[['Pct_Obese', 'ZIP']].sort_values(by=['Pct_Obese'], ascending=False)
```

```python
df_gpd['Pct_Obese'] = (df_gpd['Pct_Obese'] / 1000)
```

```python

df_gpd.plot("Pct_Obese", legend=True)
```

the map shows a higher level of obesity in the northern portion of map especially in the NW. This is also consistent with the diabetes discharge rates observed earlier showing a postive relationship between diabetes and obesity in Lake County, Illinois.
